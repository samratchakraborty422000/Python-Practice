## Decision Tree Classifier Implementation With Post Prunning And Preprunning

image.png

image.png

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
dataset=load_iris()
print(dataset.DESCR)
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
dataset.target
```

## Feature Extraction

```python
#independent and dependent variable
X=df.iloc[:,:-1]
y=dataset.target
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

image.png

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(X_train,y_train)
```

image.png

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

## Post Pruning

```python
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=2) #we can select the masxepth to perform pruning
classifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifier,filled=True)
```

```python
##prediction
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

## Performance Metrics

```python
from sklearn.metrics import accuracy_score,classification_report
score=accuracy_score(y_pred,y_test)
print("accurecy score is: ",score)
print(classification_report(y_pred,y_test))
```

post-pruning can be effectively used for smaller datasets. While pre-pruning (stopping the tree's growth early) is often preferred for larger datasets due to its efficiency

## DecisionTree Prepruning And Hyperparameter Tuning For Huge Data

image.png

```python
#we will select the best parameters based on accurecy score via GridSearchCV
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
classifier=DecisionTreeClassifier()
clf=GridSearchCV(classifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred=clf.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print("accurecy score is: ",score)
print(classification_report(y_pred,y_test))
```

# Decision Tree Regressor

```python
##California House Pricing Dataset
from sklearn.datasets import fetch_california_housing
california=fetch_california_housing()
```

```python
print(california.DESCR)
```

```python
df=pd.DataFrame(california.data,columns=california.feature_names)
df['Target']=california.target
```

```python
df.shape
```

## Feature Extraction

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
```

image.png

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_pred,y_test)
```

```python
score
```

## Hyperparameter Tuning


```python
parameter={
 'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
regressor=DecisionTreeRegressor()
```

```python
regressorcv=GridSearchCV(regressor,param_grid=parameter,cv=2,scoring='neg_mean_squared_error')
```

```python
regressorcv.fit(X_train,y_train)
```

```python
regressorcv.best_params_
```

```python
regressor=DecisionTreeRegressor(criterion='squared_error',max_depth=10,max_features='log2',splitter='best')
```

```python
regressor.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(regressor,filled=True)
plt.show()
```

```python
y_pred=regressorcv.predict(X_test)
```

```python
r2_score(y_pred,y_test)
```
